```python
# Product URL
prod_url = []
url= driver.find_elements_by_xpath('//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
for i in url:
    prod_url.append(i.get_attribute('href'))
```

```python
# Second page
try:
    next_btn1= driver.find_element_by_xpath('//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
    driver.get(next_btn1.get_attribute('href'))
    time.sleep(5)
    url= driver.find_elements_by_xpath('//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in url:
        prod_url.append(i.get_attribute('href'))
except NoSuchElementException :
        pass
```

```python
len(prod_url)
```

```python
# Third page
try:
    next_btn1= driver.find_element_by_xpath('//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
    driver.get(next_btn1.get_attribute('href'))
    time.sleep(5)
    url= driver.find_elements_by_xpath('//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in url:
        prod_url.append(i.get_attribute('href'))
except NoSuchElementException :
        pass
```

```python
len(prod_url)
```

```python
pn= []
bn=[]
pp=[]
pr=[]
pde=[]
ps=[]
for i in prod_url:
    driver.get(i)
    time.sleep(5)
    try:
        prod_name= driver.find_element_by_xpath('//*[@id="title_feature_div"]')
        pn.append(prod_name.text)
    except NoSuchElementException :
        pn.append('-')
    try:
        brand_name = driver.find_element_by_xpath('//*[@id="productOverview_feature_div"]/div/table/tbody/tr[1]/td[2]/span')
        bn.append(brand_name.text)
    except NoSuchElementException :
        bn.append('-')
    try:
        prod_price= driver.find_element_by_xpath('//*[@id="corePrice_feature_div"]/div/span/span[2]/span[2]')
        pp.append(prod_price.text)
    except NoSuchElementException :
        pp.append('-')   
    try:
        prod_rep= driver.find_element_by_xpath('//*[@id="RETURNS_POLICY"]/span/div[2]/a')
        pr.append(prod_rep.text)
    except NoSuchElementException :
        pr.append('-') 
    try:    
        prod_del= driver.find_element_by_xpath('//*[@id="mir-layout-DELIVERY_BLOCK-slot-DELIVERY_MESSAGE"]/b')
        pde.append(prod_del.text)
    except NoSuchElementException :
        pde.append('-') 
    try:
        prod_stock= driver.find_element_by_xpath('//*[@id="availability"]/span')
        ps.append(prod_stock.text)
    except NoSuchElementException :
        ps.append('-') 
      
```

```python
print(len(pn))
print(len(bn))
print(len(pp))
print(len(pr))
print(len(pde))
print(len(ps))
```

```python
product_details = pd.DataFrame({'Product Name' :pn,'Product brand':bn,'Price':pp,'Return/Exchange':pr,'Expected Delivery':pde,'Availability':ps,'Product URL':prod_url})
```

```python
# Creating empty lists to store data

phone_name=[]
phone_colour=[]
phone_ram=[]
phone_storage=[]
camera_p=[]
camera_s=[]
display_size=[]
display_resolution=[]
phone_processor=[]
processor_cores=[]
battery_capacity=[]
urls=[]
```

```python
# Collecting URLs for each product on the page

for links in driver.find_elements_by_xpath("//a[@class='_1fQZEK']"):
    urls.append(links.get_attribute('href'))
```

```python
len(urls)
```

```python
# Iterating through the URL list and accessing each link to extract data    
for url in urls:
    driver.get(url)
    time.sleep(5)

    # Finding the Read More button and clicking it
    driver.find_element_by_xpath("//button[@class='_2KpZ6l _1FH0tX']").click()
    time.sleep(3)
    
    # Storing all the table rows
    model=driver.find_elements_by_xpath("//tr[@class='_1s_Smc row']")

    # Iterating through the Table rows to get required data
    for name in model:
        text1=name.find_element_by_xpath("td[1]").text # Storing text for first Table Data for each row to check 'if' condition
    
        # Extracting Model name
        try:
            if text1=='Model Name':
                phone_name.append(name.find_element_by_xpath("td[2]").text)
        except NoSuchElementException :
            phone_name.append('-')
        # Extracting Phone Color
        try:
            if text1=='Color':
                phone_colour.append(name.find_element_by_xpath("td[2]").text)
        except NoSuchElementException :
            phone_colour.append('-')

        # Extracting Display Size
        try:
            if text1=='Display Size':
                display_size.append(name.find_element_by_xpath("td[2]").text)
        except NoSuchElementException :
            display_size.append('-')

        # Extracting Display Resolution
        try:
            if text1=='Resolution':
                display_resolution.append(name.find_element_by_xpath("td[2]").text)
        except NoSuchElementException :
            display_resolution.append('-')

        # Extracting Internal Storage(ROM)
        try:
            if text1=='Internal Storage':
                phone_storage.append(name.find_element_by_xpath("td[2]").text)
        except NoSuchElementException :
            phone_storage.append('-')

        # Extracting RAM Memory
        try:
            if text1=='RAM':
                phone_ram.append(name.find_element_by_xpath("td[2]").text)
        except NoSuchElementException :
            phone_ram.append('-')

        # Extracting Primary Camera Details
        try:
            if text1=='Primary Camera':
                camera_p.append(name.find_element_by_xpath("td[2]").text)
        except NoSuchElementException :
            camera_p.append('-')

        # Extracting Battery Capacity
        try:
            if text1=='Battery Capacity':
                battery_capacity.append(name.find_element_by_xpath("td[2]").text)
        except NoSuchElementException :
            battery_capacity.append('-')
```

```python
print(len(phone_name))
print(len(phone_colour))
print(len(display_size))
print(len(display_resolution))
print(len(phone_storage))
print(len(phone_ram))
print(len(camera_p))
print(len(camera_s))
print(len(battery_capacity))
```

```python
# Creating final Data Set

mobile_details = pd.DataFrame({'phone_name' :phone_name,'phone_colour':phone_colour,'display_size':display_size,'display_resolution':display_resolution,'phone_storage':phone_storage,'phone_ram':phone_ram,'Primary camera':camera_p,'Secondary camera':camera_s,'Battery':battery_capacity})
```
