```python
#importing the requiredlibraries
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://www.dineout.co.in/delhi-restaurants')
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
title =[]
for i in soup.find_all('a',class_= "restnt-name ellipsis"):
    title.append(i.text)
title
```

```python
cuisine =[]
for c in soup.find_all('span',class_= "double-line-ellipsis"):
    cuisine.append(c.text.split('|')[1])
cuisine
```

```python
loc =[]
for l in soup.find_all('div',class_= "restnt-loc ellipsis"):
    loc.append(l.text)
loc
```

```python
rate =[]
for r in soup.find_all('div',class_= "restnt-rating rating-4"):
    rate.append(r.text)
rate
```

```python
image =[]
for e in soup.find_all('img',class_= "no-img"):
    image.append(e['data-src'])
image
```

```python
print(len(title),len(cuisine),len(loc),len(rate),len(image))
```

```python
import pandas as pd
df = pd.DataFrame({'Restaurant Name':title,'Cuisine':cuisine,'Location':loc,'Rating':rate,'Image URL':image})
```
