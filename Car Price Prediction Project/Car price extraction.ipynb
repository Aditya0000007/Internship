```python
nm=[]
km=[]
acc=[]
own=[]
fuel=[]
trans=[]
year=[]
price=[]
```

```python
for i in cars:
    try:
        driver.get(i)
        time.sleep(5)
        print(i)
        SCROLL_PAUSE_TIME = 0.5

    # Get scroll height
        last_height = driver.execute_script("return document.body.scrollHeight")

        while True:
    # Scroll down to bottom
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Wait to load page
            time.sleep(SCROLL_PAUSE_TIME)

    # Calculate new scroll height and compare with last scroll height
            new_height = driver.execute_script("return document.body.scrollHeight")
            if new_height == last_height:
                break
            last_height = new_height
        
        try:
           
            name= driver.find_elements_by_xpath('//div[@class="_187DF"]')
            nm.append(name[0].text.splitlines()[0])
        except NoSuchElementException:
            nm.append(np.nan)
        except IndexError:
            nm.append(np.nan)
        try:
            
            kilo= driver.find_elements_by_xpath('//strong[@class="_1-PH2"]')
            km.append(kilo[2].text[:-2])
        except NoSuchElementException:
            km.append(np.nan)
        except IndexError:
            km.append(np.nan)
        try:

            accd=driver.find_elements_by_xpath('//strong[@class="_1-PH2"]')
            acc.append(accd[0].text)
        except NoSuchElementException:
            acc.append(np.nan)
        except IndexError:
            acc.append(np.nan)
        try:
            
            owne=driver.find_elements_by_xpath('//strong[@class="_1-PH2"]')
            own.append(owne[1].text)
        except NoSuchElementException:
            own.append(np.nan)
        except IndexError:
            own.append(np.nan)
        try:
            
            fue=driver.find_elements_by_xpath('//strong[@class="_1-PH2"]')
            fuel.append(fue[3].text)
        except NoSuchElementException:
            fuel.append(np.nan)
        except IndexError:
            fuel.append(np.nan)
        try:
            
            tran=driver.find_elements_by_xpath('//strong[@class="_1-PH2"]')
            trans.append(tran[5].text)
        except NoSuchElementException:
            trans.append(np.nan)
        except IndexError:
            trans.append(np.nan)
        try:
            
            yr=driver.find_elements_by_xpath('//strong[@class="_1-PH2"]')
            year.append(tran[8].text)
        except NoSuchElementException:
            year.append(np.nan)
        except IndexError:
            year.append(np.nan)
        try:
            
            pr=driver.find_element_by_xpath('//strong[@class="_2yYvS Wqfj6"]')
            price.append(pr.text.splitlines()[1])
        except NoSuchElementException:
            price.append(np.nan)
        except IndexError:
            price.append(np.nan)
    except IndexError:
        pass
    continue
   
   
```

```python
# Final Dataset
dataset_final= pd.DataFrame({'Name' :nm,'kilometer ':km,'Accidential':acc,'Owner':own,'Fuel type':fuel,'Transmission type':trans,'Manufacture year':year,'Price':price})
```
